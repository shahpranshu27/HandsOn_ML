```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/homeprices.csv")
df
```

```python
plt.scatter(df['area'],df['price'], marker='+', color='red')
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
model = LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.predict([[3300]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
area_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/areas.csv")
area_df
```

```python
p = model.predict(area_df)
```

```python
area_df['prices'] = p
area_df
```

```python
area_df.to_csv("/content/drive/MyDrive/Colab Notebooks/predictions.csv")
```
