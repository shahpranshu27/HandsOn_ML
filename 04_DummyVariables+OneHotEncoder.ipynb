```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/homeprices2.csv')
df
```

## Using Dummy Variables

```python
pd.get_dummies(df.town)
```

```python
print(df.town.dtype)
```

```python
df['town'] = df['town'].astype('category')
df.dtypes
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
dummies['monroe township'] = dummies['monroe township'].replace({True: 1, False: 0})
dummies['west windsor'] = dummies['west windsor'].replace({True: 1, False: 0})
dummies['robinsville'] = dummies['robinsville'].replace({True: 1, False: 0})
dummies
```

```python
dummies['monroe township'] = dummies['monroe township'].astype(int)
dummies['west windsor'] = dummies['west windsor'].astype(int)
dummies['robinsville'] = dummies['robinsville'].astype(int)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
model = LinearRegression()
```

```python
X = final.drop('price',axis='columns')
X
```

```python
Y = final.price
Y
```

```python
model.fit(X, Y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X, Y)
```

## Using sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town', 'area']].values
y = dfle.price.values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:, 1:]
X
```

```python
model.fit(X, y)
```

```python
model.predict([[1,0,2800]])
```
