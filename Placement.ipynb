```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# /content/drive/MyDrive/Colab Notebooks/placement.csv
df = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/placement.csv")
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('cgpa')
plt.ylabel('package(in lpa)')
```

```python
X = df.iloc[:, 0:1] # input
y = df.iloc[:, -1] # output
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'], df['package'])
plt.plot(X_train, lr.predict(X_train), color="red")
plt.xlabel('cgpa')
plt.ylabel('package(in lpa)')
```

```python
m = lr.coef_ # value of m in y=mx+c
```

```python
c = lr.intercept_ # value of c in y=mx+c
```
