# **Simple Linear Regression**

Dataset taken from

https://www.kaggle.com/datasets/ravitejakotharu/salary-datacsv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r'/content/drive/MyDrive/Colab Notebooks/salary_data.csv')
x = df.iloc[:, :-1].values # gets copy of df exclude last column
y = df.iloc[:, -1].values # gets array of df in 1st column
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train) # X_train -> years_of_exp, Y_train -> salary
```

## Visualization

```python
# visualizing the training set results
plt.scatter(X_train, Y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title("salary vs experience (Training set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()

# visualizing the test set results
plt.scatter(X_test, Y_test, color="red")
plt.plot(X_test, regressor.predict(X_test), color="blue")
plt.title("salary vs experience (Test set)")
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.show()
```

```python
# Prediction of salary, for 5 years of experience
y_pred = regressor.predict(np.array([[5]]))
print(y_pred)
```

## Conclusion
For a candidate with 5 years of experience, he will get the salary of $73545.90
